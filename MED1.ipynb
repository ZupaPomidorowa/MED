```python
species = iris['species'].unique()
for species in iris['species'].unique():
    print(species)
    print(iris[iris['species'] == species].describe())
```

```python
#Iris histogram
atr_arr = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

for i, atr in enumerate(atr_arr):
    row = i // 2  
    col = i % 2 

    for species in iris['species'].unique():
        sns.distplot(iris.loc[iris.species == species, atr], label=species, ax=axes[row, col])

    axes[row, col].set_title(f"{atr}")
    axes[row, col].legend()


plt.tight_layout()
plt.show()
```

```python
#Iris wykresy pudełkowe
art_arr = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

plt.figure(figsize=(10,8), dpi= 80)

for i in range(1, 5):
    plt.subplot(2,2,i)
    sns.boxplot(x='species', y=art_arr[i-1], data=iris)
```
